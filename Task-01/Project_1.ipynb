```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset for Data Analytics - Sheet1.csv')
print("Shape:", df.shape)
```

```python
df.head()
```

```python
print("Data types:")
print(df.dtypes)
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate OrderIDs:", df['OrderID'].duplicated().sum())
```

```python
print("Full row duplicates:", df.duplicated().sum())
```

```python
change_log = []
```

```python
n_before = df['CouponCode'].isnull().sum()

df['CouponCode'] = df['CouponCode'].fillna('NONE')

change_log.append(('CR001', 'Filled missing CouponCode with NONE (no coupon used)', f'{n_before} rows fixed', 'Resolved'))
print(f"Fixed {n_before} rows")
```

```python
print(df.isnull().sum())
```

```python
n_id_dupes = df.duplicated(subset='OrderID').sum()
df = df.drop_duplicates(subset='OrderID', keep='first')

n_full_dupes = df.duplicated().sum()
df = df.drop_duplicates()

change_log.append(('CR002', 'Checked and removed duplicate OrderIDs', f'{n_id_dupes} rows removed', 'Resolved'))
change_log.append(('CR003', 'Checked and removed full-row duplicates', f'{n_full_dupes} rows removed', 'Resolved'))
print(f"Removed {n_id_dupes} ID duplicates, {n_full_dupes} full duplicates")
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

bad_dates = df['Date'].isnull().sum()
change_log.append(('CR004', 'Verified/standardized Date to ISO 8601 (YYYY-MM-DD)', f'{bad_dates} invalid dates found', 'Resolved'))
print(f"Invalid dates: {bad_dates}")
```

```python
text_cols = ['OrderID', 'CustomerID', 'Product', 'ShippingAddress',
             'PaymentMethod', 'OrderStatus', 'TrackingNumber',
             'CouponCode', 'ReferralSource']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

change_log.append(('CR005', 'Trimmed whitespace on all text columns', 'All rows', 'Resolved'))
```

```python
for col in ['UnitPrice', 'TotalPrice']:
    df[col] = df[col].round(2)

change_log.append(('CR006', 'Rounded UnitPrice and TotalPrice to 2 decimals', 'All rows', 'Resolved'))
```

```python
calc = (df['Quantity'] * df['UnitPrice']).round(2)
mismatch = (calc - df['TotalPrice']).abs() > 0.01
n_mismatch = mismatch.sum()

df.loc[mismatch, 'TotalPrice'] = calc[mismatch]

change_log.append(('CR007', 'Verified/corrected TotalPrice = Quantity x UnitPrice', f'{n_mismatch} rows corrected', 'Resolved'))
print(f"Corrected {n_mismatch} mismatches")
```

```python
assert df['OrderID'].duplicated().sum() == 0, "FAIL: duplicate OrderIDs remain"
assert df['Date'].isnull().sum() == 0, "FAIL: bad dates remain"
assert df.isnull().sum().sum() == 0, "FAIL: missing values remain"

print("✅ VERIFICATION PASSED")
```

```python
print(df.isnull().sum())
print("\nFinal shape:", df.shape)
df.head()
```

```python
df.to_csv('Cleaned_Dataset for Data Analytics - Sheet1.csv', index=False)

change_log_df = pd.DataFrame(change_log, columns=['Change ID', 'Description', 'Impact', 'Status'])
change_log_df.to_csv('change_log.csv', index=False)

print(change_log_df.to_string(index=False))
```
